```python
import pandas as pd
import numpy as np
import seaborn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
non_primed = pd.read_pickle("non_primed_no_duplicate.pickle")
primed = pd.read_pickle("primed_no_duplicate.pickle")
```

```python
primed_credits = [primed.QL1_2_loa.mean().round(2), primed.QL2_3_loa.mean().round(2), primed.QL3_2_loa.mean().round(2), primed.QL3_3_loa.mean().round(2), primed.NW1_2_loa.mean().round(2), primed.NW2_1_loa.mean().round(2), primed.NW2_3_loa.mean().round(2), primed.CR2_2_loa.mean().round(2)]
```

```python
primed_credits
```

```python
non_primed_credits = [non_primed.QL1_2_loa.mean().round(2), non_primed.QL2_3_loa.mean().round(2), non_primed.QL3_2_loa.mean().round(2), non_primed.QL3_3_loa.mean().round(2), non_primed.NW1_2_loa.mean().round(2), non_primed.NW2_1_loa.mean().round(2), non_primed.NW2_3_loa.mean().round(2), non_primed.CR2_2_loa.mean().round(2)]
```

```python
non_primed_credits
```

```python
fig, ax = plt.subplots(figsize=(8, 6), dpi=100, facecolor='w', edgecolor='k')

n_groups = 8
bar_width = 0.35
index = np.arange(n_groups)
opacity = 0.4
primed1 = plt.bar(index, primed_credits, bar_width,
                 alpha=opacity,
                 color='g',
                 label='Example Group')

non_primed2 = plt.bar(index + bar_width, non_primed_credits, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Control Group')

plt.xlabel('Envision Credits')
plt.ylabel('Scores')
plt.title('Average Credit Scores by Group Type')
plt.xticks(index + bar_width, ('QL1.2', 'QL2.3', 'QL3.2', 'QL3.3', 'NW1.2', 'NW2.1', 'NW2.3', 'CR2.2'))
lgd = plt.legend(bbox_to_anchor=[1, 1], loc='center right')
plt.tight_layout()
fig.savefig('envision-credits-8-25-2015.png', bbox_extra_artists=(lgd,), bbox_inches='tight')  # save the figure to file
plt.show()
plt.close(fig)  
```

###Adding Standard Deviation

```python
primed_credits_std = [primed.QL1_2_loa.std().round(2), primed.QL2_3_loa.std().round(2), primed.QL3_2_loa.std().round(2), primed.QL3_3_loa.std().round(2), primed.NW1_2_loa.std().round(2), primed.NW2_1_loa.std().round(2), primed.NW2_3_loa.std().round(2), primed.CR2_2_loa.std().round(2)]
```

```python
non_primed_credits_std = [non_primed.QL1_2_loa.std().round(2), non_primed.QL2_3_loa.std().round(2), non_primed.QL3_2_loa.std().round(2), non_primed.QL3_3_loa.std().round(2), non_primed.NW1_2_loa.std().round(2), non_primed.NW2_1_loa.std().round(2), non_primed.NW2_3_loa.std().round(2), non_primed.CR2_2_loa.std().round(2)]
```

```python
fig, ax = plt.subplots()

n_groups = 8
bar_width = 0.35
index = np.arange(n_groups)
opacity = 0.4
error_config = {'ecolor': '0.3'}

primed1 = plt.bar(index, primed_credits, bar_width,
                  alpha=opacity,
                  color='g',
                  yerr=primed_credits_std,
                  error_kw=error_config,
                  label='Example Group')

non_primed2 = plt.bar(index + bar_width, non_primed_credits, bar_width,
                      alpha=opacity,
                      color='b',
                      yerr=non_primed_credits_std,
                      error_kw=error_config,
                      label='Control Group')

plt.xlabel('Envision Credits')
plt.ylabel('Scores')
plt.title('Average Credit Scores by Group Type')
plt.xticks(index + bar_width, ('QL1.2', 'QL2.3', 'QL3.2', 'QL3.3', 'NW1.2', 'NW2.1', 'NW2.3', 'CR2.2'))
plt.legend(bbox_to_anchor=(1.2, 1.05))

plt.tight_layout()
plt.show()
```
